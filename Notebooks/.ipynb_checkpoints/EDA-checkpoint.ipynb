```python
df = df.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis = 1)
```

```python
sorty = df.sort_values(by=['Vote Average', 'Vote Count' ], ascending=[False, False])
```

```python
sorty
```

```python
director_counts = sorty['Director'].value_counts()
bottom_20_directors = director_counts.tail(0)
bottom_20_directors
```

```python
director_counts = sorty['Director'].value_counts()
filtered_directors = director_counts[(director_counts >= 10) & (director_counts <= 40)]
filtered_directors_df = filtered_directors.reset_index()
filtered_directors_df.columns = ['Director', 'Number of Movies']
pd.set_option('display.max_rows', None)
filtered_directors_df
```

```python
# List of specified directors
specified_directors = [
    'Petter Baiestorf', 'Katsumi Sasaki', 'Yôhei Fukuda', 'Kaosu Sumikura', 'Dustin Mills', 
    'Daisuke Morishima', 'Naoyuki Tomomatsu', 'Bill Zebub', 'William Hellfire', 'Kazushi Miki', 
    'James Balsamo', 'Alex Wesley', 'Jim Wynorski', 'Rizal Mantovani', 'Takashi Shimizu', 
    'Jirô Nagae', 'Norio Tsuruta', 'Daisuke Yamanouchi', 'Louisa Warren', 'Yozo Tainaka', 'Andrew Jones'
]

# Filter the sorty dataframe to include only the specified directors
filtered_sorty = sorty[sorty['Director'].isin(specified_directors)]

# Group by director and calculate the mean vote average
vote_averages = filtered_sorty.groupby('Director')['Vote Average'].mean().reset_index()

# Ensure the directors are in the specified order
vote_averages['Director'] = pd.Categorical(vote_averages['Director'], categories=specified_directors, ordered=True)
vote_averages = vote_averages.sort_values('Director')

# Plot the bar chart
plt.figure(figsize=(14, 8))
sns.barplot(x='Director', y='Vote Average', data=vote_averages, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Director')
plt.ylabel('Average Vote')
plt.title('Average Vote Scores for Specified Directors')
plt.show()
```

```python
# List of specified directors
specified_directors = [
    'Petter Baiestorf', 'Katsumi Sasaki', 'Yôhei Fukuda', 'Kaosu Sumikura', 'Dustin Mills', 
    'Daisuke Morishima', 'Naoyuki Tomomatsu', 'Bill Zebub', 'William Hellfire', 'Kazushi Miki', 
    'James Balsamo', 'Alex Wesley', 'Jim Wynorski', 'Rizal Mantovani', 'Takashi Shimizu', 
    'Jirô Nagae', 'Norio Tsuruta', 'Daisuke Yamanouchi', 'Louisa Warren', 'Yozo Tainaka', 'Andrew Jones'
]

# Filter the sorty dataframe to include only the specified directors
filtered_sorty = sorty[sorty['Director'].isin(specified_directors)]

# Group by director and calculate the mean popularity
popularity_averages = filtered_sorty.groupby('Director')['Popularity'].mean().reset_index()

# Ensure the directors are in the specified order
popularity_averages['Director'] = pd.Categorical(popularity_averages['Director'], categories=specified_directors, ordered=True)
popularity_averages = popularity_averages.sort_values('Director')

# Plot the bar chart
plt.figure(figsize=(14, 8))
sns.barplot(x='Director', y='Popularity', data=popularity_averages, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Director')
plt.ylabel('Popularity')
plt.title('Popularity Scores for Specified Directors')
plt.show()
```
