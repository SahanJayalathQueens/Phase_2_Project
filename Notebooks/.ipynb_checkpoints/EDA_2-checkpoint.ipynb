```python
filtered_countries = ['United States of America', 'Canada']
filtered_df = df[(df['Vote Count'] >= 100) & (df['Production Countries'].isin(filtered_countries))]

average_vote = filtered_df.groupby('Director')['Vote Average'].mean().reset_index()

sorted_average_vote = average_vote.sort_values(by='Vote Average', ascending=False)

print("DataFrame with average Vote Average by Director (with more than 100 votes and in specified countries):")
sorted_average_vote
```

```python
top_20_directors = sorted_average_vote.head(30)

plt.figure(figsize=(12, 8))
sns.barplot(x='Director', y='Vote Average', data=top_20_directors, palette='inferno', hue='Director', dodge=False)

plt.xticks(rotation=45, ha='right')

plt.title('Top 20 Directors by Average Vote (with more than 100 votes in USA)')
plt.xlabel('Director')
plt.ylabel('Average Vote')

plt.show()
```

```python
directors_with_3_entries = df.groupby('Director').filter(lambda x: len(x) >= 3)

production_countries_filter = ['United States of America', 'Canada', 'United States of America', 'United States of America', 'Canada', 'United States of America']
filtered_df = directors_with_3_entries[directors_with_3_entries['Production Countries'].isin(production_countries_filter)]

average_popularity = filtered_df.groupby('Director')['Popularity'].mean().reset_index()
average_popularity.rename(columns={'Popularity': 'Average Popularity Score'}, inplace=True)
average_popularity = average_popularity.sort_values(by='Average Popularity Score', ascending=False)

average_popularity
```

```python
top_20_directors = average_popularity.head(30)

# Create a bar plot using Seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='Director', y='Average Popularity Score', data=top_20_directors, palette='viridis', hue='Director', dodge=False)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add title and labels
plt.title('Top 20 Directors by Average Popularity Score')
plt.xlabel('Director')
plt.ylabel('Average Popularity Score')

# Show the plot
plt.show()
```

```python
nu_df = df.groupby('Director')['Budget','Revenue'].reset_index()
```

```python
nu_df = df[['Director', 'Budget', 'Revenue']]
nu_df 
```

```python
nu_no_nan = nu_df.dropna()
nu_no_nan 
```

```python
nu_no_zeros = nu_no_nan[(nu_no_nan['Budget'] != 0) & (nu_no_nan['Revenue'] != 0)]
```

```python
nu_no_zeros.info()
```

```python
cash_cow = nu_no_zeros.groupby('Director').agg(
    Budget=('Budget', 'mean'),
    Revenue=('Revenue', 'mean')
).reset_index()

# Step 2: Calculate ROI
cash_cow['ROI'] = cash_cow['Budget'] - cash_cow['Revenue']

# Display the new DataFrame
cash_cow
```

```python
cash_cow_sorted = cash_cow.sort_values(by='ROI', ascending=False)
cash_cow_sorted
```

```python
cash_cow_sorted.info()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Assuming cash_cow is already created and contains the necessary data

# Step 1: Prepare the data for linear regression
X = cash_cow[['Budget']]  # Predictor variable
y = cash_cow['Revenue']    # Response variable

# Step 2: Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Step 3: Make predictions
cash_cow['Predicted_Revenue'] = model.predict(X)

# Step 4: Calculate the slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Display slope and intercept
print(f"Slope: {slope}, Intercept: {intercept}")

# Step 5: Visualize the results with a scatter plot and regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(data=cash_cow, x='Budget', y='Revenue', label='Actual Revenue')
plt.plot(cash_cow['Budget'], cash_cow['Predicted_Revenue'], color='red', label='Regression Line')
plt.title('Linear Regression: Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error


# Step 1: Predictions have already been made
predicted_revenue = cash_cow['Predicted_Revenue']
actual_revenue = cash_cow['Revenue']

# Step 2: Calculate R-squared
r_squared = r2_score(actual_revenue, predicted_revenue)

# Step 3: Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(actual_revenue, predicted_revenue)

# Step 4: Calculate Mean Squared Error (MSE)
mse = mean_squared_error(actual_revenue, predicted_revenue)

# Display the metrics
print(f"R-squared: {r_squared:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
```

```python
import pandas as pd
import statsmodels.api as sm

# Assuming cash_cow is already created with Budget, Revenue, and ROI

# Step 1: Prepare the data for regression
# You can choose ROI as the dependent variable
X = cash_cow[['Budget', 'Revenue']]  # Independent variables
y = cash_cow['ROI']  # Dependent variable

# Add a constant to the predictor variables (for the intercept)
X = sm.add_constant(X)

# Step 2: Fit the multiple linear regression model
model = sm.OLS(y, X).fit()

# Step 3: Get the summary of the model
summary = model.summary()

# Display the summary, which includes p-values for each predictor
summary
```

```python
from sklearn.linear_model import LinearRegression
from matplotlib.ticker import FuncFormatter
from sklearn.metrics import r2_score

# Assuming cash_cow is already created and contains the necessary data

# Step 1: Prepare the data for linear regression
X = cash_cow[['Budget']]  # Predictor variable
y = cash_cow['Revenue']    # Response variable

# Step 2: Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Step 3: Make predictions
cash_cow.loc[:, 'Predicted_Revenue'] = model.predict(X)  # Use .loc to avoid warning

# Step 4: Calculate the slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

# Display slope and intercept
print(f"Slope: {slope}, Intercept: {intercept}")

# Step 5: Visualize the results with a scatter plot and regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(data=cash_cow, x='Budget', y='Revenue', label='Actual Revenue')
plt.plot(cash_cow['Budget'], cash_cow['Predicted_Revenue'], color='red', label='Regression Line')

# Format the axes to show values in millions
def million_formatter(x, pos):
    return f'${x * 1e-6:.0f}M'  # Format to millions

plt.gca().xaxis.set_major_formatter(FuncFormatter(million_formatter))
plt.gca().yaxis.set_major_formatter(FuncFormatter(million_formatter))

# Zooming in on the bottom right corner
plt.xlim(0, 90000000)  # Adjust based on your data
plt.ylim(0, 150000000)  # Adjust based on your data

# Calculate R-squared
actual_revenue = cash_cow['Revenue']
predicted_revenue = cash_cow['Predicted_Revenue']
r_squared = r2_score(actual_revenue, predicted_revenue)

plt.title('Linear Regression: Budget vs Revenue')
plt.xlabel('Budget (in millions)')
plt.ylabel('Revenue (in millions)')
plt.legend()
plt.grid()
plt.show()

print(f"R-squared: {r_squared:.4f}")
```
