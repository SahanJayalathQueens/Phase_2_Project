```python
import requests
import pandas as pd

api_key = '95408ea58dbcf11b6cfd9e44971ff7a4'

url = 'https://api.themoviedb.org/3/discover/movie'

params = {
    'api_key': api_key,
    'with_genres': '27',       # Genre ID for Horror
    'primary_release_date.gte': '1996-01-01',  # Starting from Jan 1, 1996
    'primary_release_date.lte': '1997-12-31',  # Until Dec 31, 1997
    'language': 'en-US',       # Results in English
    'sort_by': 'popularity.desc',  # Sort by popularity
    'page': 1                  # Pagination, start with page 1
}

def get_movies(url, params):
    response = requests.get(url, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        return None

movies_data = get_movies(url, params)

movies_list = []

if movies_data:
    # Loop through each movie in the results
    for movie in movies_data['results']:
        # Extract relevant movie information
        movie_info = {
            'Title': movie['title'],
            'Release Date': movie['release_date'],
            'Popularity': movie['popularity'],
            'Vote Average': movie['vote_average'],
            'Overview': movie['overview']
        }
        movies_list.append(movie_info)

    movies_df = pd.DataFrame(movies_list)

    print(movies_df)
else:
    print("No data found.")
```
